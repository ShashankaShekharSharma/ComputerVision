```python
# Use of MediaPipe for Hand Tracking using CPU
# installed opencv-python, mediapipe
import cv2
import time
import mediapipe as mp

cap = cv2.VideoCapture(0)
mpHands = mp.solutions.hands 
hands = mpHands.Hands(False)
mpDraw = mp.solutions.drawing_utils
pTimer = 0
cTimer = 0

while True:
    success, img = cap.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)

    #print(results.multi_hand_landmark)
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                h,w,c = img.shape
                cx, cy = int(lm.x*w), int(lm.y*h)
                print(id, cx, cy)
                if id == 0:
                    cv2.circle(img, (cx,cy), 15, (255,0,255), cv2.FILLED)
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)

    cTimer = time.time()
    fps = 1/(cTimer-pTimer)
    pTimer = cTimer

    cv2.putText(img, str(int(fps)), (10,70), cv2.FONT_HERSHEY_PLAIN, 3, (255,0,255), 3)

    cv2.imshow("Image",img)
    cv2.waitKey(1)
```

## Average Precision

**Confusion Matrix**
The confusion matrix is a tool used to understand the performance of a classification model, which is a type of machine learning model that predicts which category or class an item belongs to

Imagine you're a doctor diagnosing patients as either having a disease (Positive) or not having it (Negative) based on some medical test.

When the test results come in, there are four possible outcomes for each patient:

- True Positive (TP): The test correctly predicts that the patient has the disease.
- True Negative (TN): The test correctly predicts that the patient does not have the disease.
- False Positive (FP): The test incorrectly predicts that the patient has the disease (also known as a "false alarm").
- False Negative (FN): The test incorrectly predicts that the patient does not have the disease (missed diagnosis).

The confusion matrix for a binary classifier can be visualized as:

$$
\begin{array}{c|c|c}
\text{} & \text{Predicted Positive} & \text{Predicted Negative} \\
\hline
\text{Actual Positive} & \text{True Positive (TP)} & \text{False Negative (FN)} \\
\hline
\text{Actual Negative} & \text{False Positive (FP)} & \text{True Negative (TN)} \\
\end{array}
$$

Using the confusion matrix, you can calculate various performance metrics:

- **Accuracy**: This measures how often the model is correct.
  $$
  \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
  $$

- **Precision**: This tells us how many of the predicted positive cases were actually positive.
  $$
  \text{Precision} = \frac{TP}{TP + FP}
  $$

- **Recall (Sensitivity)**: This tells us how many of the actual positive cases were correctly predicted.
  $$
  \text{Recall} = \frac{TP}{TP + FN}
  $$

- **F1 Score**: This is the harmonic mean of Precision and Recall, providing a single metric that balances both.
  $$
  \text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
  $$




2. **Mean Average Precision (mAP)**

Mean Average Precision (mAP) is a metric used to evaluate the performance of object detection models, which are a type of machine learning model designed to locate and classify objects within images. It combines precision and recall across different thresholds and classes, providing a comprehensive measure of a model's accuracy.

To understand mAP, we need to break it down into its core components:

1. **Precision**: Precision is the ratio of correctly predicted positive observations to the total predicted positives. It answers the question, "Of all the objects detected, how many were correct?"

   $$
   \text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Positives (FP)}}
   $$

2. **Recall**: Recall is the ratio of correctly predicted positive observations to all observations in the actual class. It answers the question, "Of all the actual objects, how many were detected correctly?"

   $$
   \text{Recall} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FN)}}
   $$

### Precision-Recall Curve

The performance of an object detection model can be visualized using a precision-recall curve, which plots precision against recall at various threshold settings. A high area under the curve represents both high recall and high precision, indicating good performance.

### Average Precision (AP)

Average Precision (AP) summarizes the precision-recall curve into a single value, taking the average precision over all recall values from 0 to 1. This involves calculating the precision at different recall levels and then averaging them. The AP is typically computed for each class in a multi-class problem.

$$
\text{AP} = \int_0^1 \text{Precision}(\text{Recall}) \, d\text{Recall}
$$

### Mean Average Precision (mAP)

Mean Average Precision (mAP) is the average of the APs over all classes. If we have \(N\) classes, the mAP is calculated as follows:

$$
\text{mAP} = \frac{1}{N} \sum_{i=1}^{N} \text{AP}_i
$$

Where \(\text{AP}_i\) is the average precision for class \(i\).

### Steps to Calculate mAP

1. **Compute Precision and Recall**: For each detected object, compare it with the ground truth to determine if it is a true positive or false positive. Calculate precision and recall for various threshold levels.

2. **Plot Precision-Recall Curve**: For each class, plot the precision-recall curve.

3. **Calculate Average Precision (AP)**: Integrate the area under the precision-recall curve to get the AP for each class.

4. **Compute Mean Average Precision (mAP)**: Average the APs across all classes to get the mAP.

### Why mAP is Important

mAP is a comprehensive metric for evaluating object detection models because it considers both the precision and recall across different thresholds and for multiple classes. It provides a balanced measure of a model's ability to accurately detect objects (precision) and to detect as many objects as possible (recall).

By optimizing for mAP, we ensure that the model performs well across various scenarios and object classes, making it a reliable metric for real-world applications of object detection.

# The basic steps involved in digital image processing are:
- _Image acquisition_: This involves capturing an image using a digital camera or scanner, or importing an existing image into a computer.
- _Image enhancement_: This involves improving the visual quality of an image, such as increasing contrast, reducing noise, and removing artifacts.
- _Image restoration_: This involves removing degradation from an image, such as blurring, noise, and distortion.
- _Image segmentation_: This involves dividing an image into regions or segments, each of which corresponds to a specific object or feature in the image.
- _Image representation and description_: This involves representing an image in a way that can be analyzed and manipulated by a computer, and describing the features of an image in a compact and meaningful way.
- _Image analysis_: This involves using algorithms and mathematical models to extract information from an image, such as recognizing objects, detecting patterns, and quantifying features.
- _Image synthesis and compression_: This involves generating new images or compressing existing images to reduce storage and transmission requirements.
Digital image processing is widely used in a variety of applications, including medical imaging, remote sensing, computer vision, and multimedia.
